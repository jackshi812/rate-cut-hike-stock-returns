# Tech Industry Data


```python
import yfinance as yf
import pandas as pd

# Tech Industry
tickers = ['AAPL', 'MSFT', 'NVDA', 'CRM', 'ORCL','AVGO','INTC','AMD','CSCO','MU']

# data
data = yf.download(tickers, start='2014-01-01', end='2024-12-31', auto_adjust=True)

# change format
data = data.swaplevel(axis=1)  
data.columns = [f"{ticker}_{col}" for ticker, col in data.columns]  

# 重置索引
data = data.reset_index()

# CSV
data.to_csv('Tech_Stocks_prices 2014_2024.csv', index=False)

print(data.head(3))
print(f"\nTotal columns: {data.shape[1]}, Total rows: {data.shape[0]}")

data.info()
```

# Tech Industry Return

```python
# Tech Industry
tickers = ['AAPL', 'MSFT', 'NVDA', 'CRM', 'ORCL','AVGO','INTC','AMD','CSCO','MU']

# Data from 2014-2024
data = yf.download(tickers, start='2014-01-01', end='2024-12-31', auto_adjust=True)

# Row=各Ticker，Column=日期
close = data['Close']

# Close→Close 日收益率
ret = close.pct_change().dropna()

# 
ret.columns = [c + '_ret' for c in ret.columns]

# 横向是各公司的日收益，纵向是日期
ret.reset_index().to_csv('Tech_Stocks_Returns 2014_2024.csv', index=False)

print(ret.head())
print(f'Rows (trading days): {ret.shape[0]}, Columns (tickers): {ret.shape[1]}')
```

# Finance Industry

```python
# 金融板块 5 家公司
tickers_fin = ['BRK-B', 'JPM', 'V','BAC','MA', 'WFC','C', 'GS', 'MS','AXP']
# Data from 2014-2024
data_fin = yf.download(tickers_fin, start='2014-01-01', end='2024-12-31', auto_adjust=True)

# Row=各Ticker，Column=日期
data_fin = data_fin.swaplevel(axis=1)
data_fin.columns = [f"{ticker}_{col}" for ticker, col in data_fin.columns]
data_fin = data_fin.reset_index()

# CSV
data_fin.to_csv('Financial_Stocks_prices 2014_2024.csv', index=False)

print(data_fin.head())
print(f"Rows: {data_fin.shape[0]}, Columns: {data_fin.shape[1]}")
```

```python
# 取收盘价
close_fin = yf.download(tickers_fin, start='2014-01-01', end='2024-12-31', auto_adjust=True)['Close']

# Close-to-Close日收益率
ret_fin = close_fin.pct_change().dropna()

# 
ret_fin.columns = [f"{c}_ret" for c in ret_fin.columns]

# reset index
ret_fin.reset_index().to_csv('Financial_Stocks_Returns 2014_2024.csv', index=False)

print(ret_fin.head())
print(f"Rows: {ret_fin.shape[0]}, Columns: {ret_fin.shape[1]}")
```
